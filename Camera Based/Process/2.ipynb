```python
import cv2
from ultralytics import YOLO
import numpy as np

# Inisialisasi model YOLO dengan path ke model yang telah dilatih
model = YOLO(r'D:\OneDrive\Bangkit Academy\Project\AR\Test\best.pt')

# Mulai menangkap video dari webcam
cap = cv2.VideoCapture(0)

# Membaca gambar baju dengan memastikan gambar tersebut memiliki kanal alpha
baju = cv2.imread('test3.png', cv2.IMREAD_UNCHANGED)

if not cap.isOpened():
    print("Error: Camera is not available.")
    exit()

while True:
    # Membaca frame dari video
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break

    # Melakukan deteksi menggunakan model
    results = model(frame)

    # Mendapatkan frame dengan annotasi hasil deteksi
    annotated_frame = results[0].plot()
    boxes = results[0].boxes.xyxy.tolist()

    if len(boxes) > 0 and baju.shape[2] == 4:  # Pastikan baju memiliki kanal alpha
        x1, y1, x2, y2 = map(int, boxes[0])  # Konversi koordinat ke integer
        width = x2 - x1 + 100
        height = (y2 - y1) * 8  # Tinggi yang diperbesar
        resized_baju = cv2.resize(baju, (width, height))

        # Membuat canvas transparan yang lebih besar
        canvas_height = max(frame.shape[0], y1 + height)
        canvas_width = max(frame.shape[1], x1 + width)
        canvas = np.zeros((canvas_height, canvas_width, 4), dtype=np.uint8)  # 4 untuk RGBA
        
        # Tempatkan frame asli ke dalam canvas (jika frame tidak transparan, buat salinannya yang transparan)
        frame_rgba = cv2.cvtColor(frame, cv2.COLOR_BGR2BGRA)  # Tambahkan alpha channel ke frame
        canvas[:frame.shape[0], :frame.shape[1], :] = frame_rgba

        # Tempatkan baju di canvas
        canvas_roi = canvas[y1:y1+height, x1:x1+width]
        roi_bg = cv2.bitwise_and(canvas_roi, canvas_roi, mask=cv2.bitwise_not(resized_baju[:, :, 3]))
        baju_fg = cv2.bitwise_and(resized_baju[:, :, :4], resized_baju[:, :, :4])  # Gunakan semua 4 channel
        canvas[y1:y1+height, x1:x1+width] = cv2.add(roi_bg, baju_fg)
        
        # Gunakan canvas sebagai frame untuk ditampilkan
        frame = canvas

    # Menampilkan frame di window
    cv2.imshow('YOLO Detection', frame)

    # Tekan 'q' untuk keluar dari loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Bebaskan resource dan tutup window
cap.release()
cv2.destroyAllWindows()
```
