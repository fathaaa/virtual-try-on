```python
from PIL import Image

# Membuka citra asli dan mask
original_image = Image.open("result.jpg")  # Path ke citra asli
mask_image = Image.open("mask_image.png")  # Path ke citra mask

# Mendapatkan ukuran citra asli
original_width, original_height = original_image.size

# Resize mask untuk memiliki ukuran yang sama dengan citra asli
resized_mask = mask_image.resize((original_width, original_height), Image.ANTIALIAS)

# Menyimpan atau menampilkan mask yang telah di-resize
resized_mask.save("path_to_resized_mask.png")
resized_mask.show()
```
